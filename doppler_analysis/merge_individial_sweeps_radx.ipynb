```python
%%time
import warnings
warnings.filterwarnings("ignore")
import glob, os, sys
import shutil
```

```python
data_dir = "/depot/dawson29/data/Projects/PERiLS/obsdata/2022/Illinois_Mobile_Radar/IOP1/COW1/"
```

```python
files = sorted(glob.glob(os.path.join(data_dir, "*_v1_*_SUR.nc")))
len(files)
```

```python
# force the if statement below to be True on first run
el = 99999999
curr_dir_list = []
for f in files:
    new_el = float(f.split("_")[-2].replace("el", ""))
    if new_el < el:
        # store new dir name
        curr_dir = f.split(".")[1]
        curr_dir_list.append(os.path.join(data_dir,curr_dir))
        # create directory
        os.makedirs(os.path.join(data_dir,curr_dir), exist_ok=True)
    # store new el
    el = new_el
    # move file
    shutil.copy(f"{f}", f"{data_dir}{os.sep}{curr_dir}{os.sep}{f.split(os.sep)[-1]}")
```

```python
new_dir = os.path.join(data_dir, "merged")
new_dir
```

```python
import subprocess
```

```python
for path in curr_dir_list:
#     args2 = ['echo',"-f", '$file/*', "-outdir", new_dir]
    args2 = ['RadxConvert', '-ag_all', '-f', path, "-outdir", new_dir]
    subprocess.Popen(args2)
#     !echo -f $file/* -outdir ./
```

```python
radar_files = sorted(glob.glob("20220322/*"))
len(radar_files)
```

```python
import xarray as xr
```

```python
def radar_info(file=None):
    ds = xr.open_dataset(file)
    print("Time: {}".format(ds.time.values))
    print("Volume Number: {}".format(ds.volume_number.values))
    print("Fixed Angle: {}".format(ds.fixed_angle.values))
    print("Sweep Number: {}".format(ds.sweep_number.values))
    print("Range: {}".format(ds.range.values.max()/1e3))
    print("Lat: {}, Lon: {}".format(ds.latitude.values, ds.longitude.values))
```

```python
for file in radar_files:
    radar_info(file=file)
```
