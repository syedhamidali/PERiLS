```python
import pyart
import numpy as np
from pyart.retrieve import get_freq_band
from matplotlib import pyplot as plt
import glob, os
```

```python
def align_field(field):
    values, counts = np.unique(field['data'], return_counts=True)
    c_value = values[np.argmax(counts)]
    field['data'] = np.array([c_value])
    return field

def align_radar_coords(radar):
    radar.longitude = align_field(radar.longitude)
    radar.latitude = align_field(radar.latitude)
    radar.altitude = align_field(radar.altitude)
    radar.altitude_agl = align_field(radar.altitude_agl)
    return radar
```

```python
def filter_radar(radar, vel_field="VEL_F", refl_field="DBZHCC_F", 
                 ncp_field="NCP", rhv_field="RHOHV", phi_field="PHIDP"):
    '''Remove noise based on velocity texture and mask all the fields'''
    # Drop some fields
    fields_to_drop = ["DBMHC", "DBMVC", "VS", "VS_F", "VL", "VL_F", "DBZHCC"]
    for field in fields_to_drop:
        if field in radar.fields:
            del radar.fields[field]
        
#     texture = pyart.retrieve.calculate_velocity_texture(radar, 
#                                                         vel_field=vel_field, 
#                                                         wind_size=5, 
#                                                         check_nyq_uniform=False)
#     radar.add_field('VT',texture,replace_existing=True)
    # create gatefilter
    gf = pyart.filters.GateFilter(radar)
    BAND = get_freq_band(radar.instrument_parameters['frequency']['data'])
    if BAND == "X":
        vt = 5
    elif BAND == "C":
        vt = 40
#     gf.exclude_above('VT', vt)
    gf.exclude_outside(refl_field, -30, 100)
    gf.include_inside(refl_field, )
    gf.include_above(refl_name, refl_thresh)
    gf.exclude_below("SNRHC", -3)
    gf_despeckeld = pyart.correct.despeckle_field(radar, refl_field, gatefilter=gf)
    corr_ZH = radar.fields[refl_field].copy()
    ZH_array = np.ma.masked_where(gf_despeckeld.gate_included == False, radar.fields[refl_field]['data'])
    corr_ZH['data'] = ZH_array
    radar.add_field('DBZH',corr_ZH,replace_existing=True)
    
    # get the mask
    mask = np.ma.getmask(radar.fields['DBZH']['data'])
    radar.scan_type = b'ppi'
    # iterate through remaining fields
    skip_fields = ["DBZHC", "VEL"]
    for field in radar.fields.keys():
        if any(field == skip_field for skip_field in skip_fields):
            continue
        # mask the field
        radar.fields[field]['data'] = np.ma.masked_where(mask, radar.fields[field]['data'])
    return radar
```

```python
def dealiase(radar, vel_name = "VEL_F"):
    try: 
        gatefilter = pyart.correct.GateFilter(radar)
        corr_vel   = pyart.correct.dealias_region_based(
            radar, vel_field=vel_name, keep_original=False, gatefilter = gatefilter)
        radar.add_field(vel_name, corr_vel, replace_existing=True)
    except:
        None
```

```python
def despeck(radar, field, gf):
    ''''''
    gf_despeckeld = pyart.correct.despeckle_field(radar, field, gatefilter=gf)
    corr_ZH = radar.fields[field].copy()
    ZH_array = np.ma.masked_where(gf_despeckeld.gate_included == False, radar.fields[field]['data'])
    smooth_data = scipy.ndimage.median_filter(ZH_array, 3)
    corr_ZH['data'] = smooth_data
    radar.add_field(str(field)+"_C", corr_ZH, replace_existing=True)
    return radar
```

```python
def mask_radar(radar, moment, gatefilter, masking=True):
    if masking:
        gatefilter=pyart.correct.despeckle_field(radar, moment, gatefilter=None, size= 15)
        mom_masked= np.ma.masked_where(gatefilter.gate_included == False, moment)
    else: 
        mom_masked=moment
    radar.add_field(str(moment)+"_C", mom_masked, replace_existing=True)
    return radar

def GateFilter(radar, refl_field, refl_thresh, vel_field, ncp_field, ncp_thresh=0.4):
    gatefilter = pyart.filters.GateFilter(radar)
    texture = pyart.retrieve.calculate_velocity_texture(radar, 
                                                        vel_field=vel_field, 
                                                        wind_size=3, 
                                                        check_nyq_uniform=False)
    radar.add_field('VT',texture,replace_existing=True)
    radar=mask_radar(radar, "VEL_C", gatefilter, masking=False)
    gatefilter.exclude_above('VT', 20)
    gatefilter.include_above(refl_field, refl_thresh)
    VELHC = scipy.ndimage.filters.median_filter(vel_field, 3)
    radar=mask_radar(radar, VELHC, gatefilter, masking=True)
    gatefilter_ref = pyart.filters.moment_based_gate_filter(radar, ncp_field=ncp_field, min_ncp=ncp_thresh)
    radar=mask_radar(radar, moment, gatefilter_ref, True)
    return radar, gatefilter, 
```

```python
def radar_fields_prep(config, rfile, radar_type, sweep_id, moment='From_config'):
    def mask(rfile, gatefilter, moment, new_mom_name, masking=True):
        if masking==True:
            #  gatefilter=pyart.correct.despeckle_field(rfile, moment, gatefilter=gatefilter, size= 15)
            mom_masked= np.ma.masked_where(gatefilter.gate_included == False, moment)
        else: mom_masked=moment
        rfile.add_field(new_mom_name, {'data': mom_masked}, replace_existing=True)
        return rfile

    gatefilter = pyart.filters.GateFilter(rfile)
    texture_feild= pyart.retrieve.calculate_velocity_texture(rfile, vel_field=vel_name, wind_size=3)
    rfile=mask(rfile, gatefilter, texture_feild['data'], 'vel_texture', masking=False)
    gatefilter.exclude_above('vel_texture', 3)
    gatefilter.include_above(refl_name, refl_thresh)

    vel_field = rfile.fields[vel_name]['data']
    smooth_data = scipy.ndimage.filters.median_filter(vel_field, 3)
    #  smooth_data = scipy.ndimage.filters.median_filter(vel_field, 3)
    #  smooth_data_ma = np.ma.masked_where(np.ma.getmask(vel_field), smooth_data)
    rfile=mask(rfile, gatefilter, smooth_data, 'vel_smooth', masking =True)
#
    if moment == 'From_config': mom = config.r_mom
    else: mom = [moment]
    for m in mom:
        print(m)

        if m in ['vel', 'refl', 'specw']:
            if m == 'vel':
                orig_field_name, new_feild_name, gfilter = vel_name, 'vel_fix', gatefilter
            if m == 'refl':
                if radar_type != 'WSR':
                    gatefilter_refl= pyart.filters.moment_based_gate_filter(rfile, ncp_field= ncp_name, min_ncp= ncp_thresh)
                else:
                    gatefilter_refl=gatefilter
                orig_field_name, new_feild_name, gfilter = refl_name, 'refl_fix', gatefilter_refl
            if m == 'specw': orig_field_name, new_feild_name, gfilter = specw_name, 'specw_fix', gatefilter
            orig_field = rfile.fields[orig_field_name]['data']
            rfile=mask(rfile, gfilter, orig_field, new_feild_name, masking=True)

        if m in ['vel_despeck', 'refl_despeck']:
            if m == 'refl_despeck': orig_field_name, new_feild_name= 'refl_fix', 'refl_despeck'
            elif m == 'vel_despeck': orig_field_name, new_feild_name= 'vel_fix', 'vel_despeck'
            speckeled_gatefilter = pyart.correct.despeckle_field(rfile, field=orig_field_name, threshold=-100,
                                                                     gatefilter=gatefilter, size=30)
            rfile=mask(rfile, speckeled_gatefilter, rfile.fields[orig_field_name]['data'], 'refl_despeck')
```

```python
def plot_radar(radar, refl_field=refl_field, vel_field=vel_field, gatefilter=gatefilter, corrected=True):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
    display = pyart.graph.RadarMapDisplay(radar)
    if corrected:
        refl_field_corr = '{}_C'.format(refl_field)
        vel_field_corr = '{}_C'.format(vel_field)
        
    display.plot_ppi(refl_field_corr, vmin=-10., vmax=70., 
                     cmap='pyart_HomeyerRainbow', gatefilter=gatefilter, ax=ax1)

    display.plot_ppi(vel_field_corr, vmin=-35., vmax=35., 
                         cmap='pyart_NWSVel', gatefilter=gatefilter, ax=ax2)

plot_radar(radar, corrected=True)
```

```python
def velocity_dealias(radar):
    # Set the nyquist
    nyq = radar.instrument_parameters['nyquist_velocity']['data'][0]
    velocity_dealiased = pyart.correct.dealias_region_based(radar, vel_field=vel_field,
                                                            nyquist_vel=nyq, centered=True)
    radar.add_field('corrected_velocity', velocity_dealiased, replace_existing=True)   
    return radar
```
