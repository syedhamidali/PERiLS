```python
gridded = "/depot/dawson29/data/Projects/PERiLS/obsdata/2022/GRID/IOP2/"
file_list = sorted(glob.glob(os.path.join(gridded, "NOXP/*nc")))

large_ds = []
for file in file_list:
    ds = xr.open_dataset(file).isel(z=4)
    large_ds.append(ds.dims) #[247, 305, 600]
#     if ds.dims['x'] == 600:
#         large_ds.append(ds)
```

```python
large_ds
```

```python
ds = xr.concat(large_ds, dim='time')
```

```python
slider = widgets.IntSlider(value=0, min=0, max=ds.time.size-1, step=1, description='Time')

# Define a function to update the plot based on the slider value
def update_plot(field='REF', time_index=0):
    if field.lower().startswith(('re', 'db')):
        cmap = HomeyerRainbow
        vmin = -10
        vmax = 80
    elif field.lower().startswith(('v')):
        cmap = NWSVel
        vmin = -70
        vmax = 70
    else:
        cmap = 'turbo'
        vmin = ds[field][0].min().values
        vmax = ds[field][0].max().values
    
    fig, ax = plt.subplots()
    ds[field].isel(time=time_index).plot(ax=ax, vmin=vmin, vmax=vmax, cmap=cmap)
    plt.show()
```

```python
# Create the interactive plot using the interact function
widgets.interact(update_plot, field=['REF', 'VEL', 'ZDR', 'RHO'], time_index=slider)
```
